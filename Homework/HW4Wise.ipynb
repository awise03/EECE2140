(Graded) Let ùêÄ be an ùëö-by-ùëõ matrix of integer numbers. Write a Python function that accepts
matrix ùêÄ as its input and returns the elements that are common across all rows of the matrix.
For example, if the 6-by-4 matrix ùêÄ is defined as
A = [[7 1 5 6]
[2 6 1 1]
[6 1 7 2]
[6 6 3 1]
[5 5 6 1]
[3 6 7 1]]
then, the function should return 1 and 6 as the common elements between the rows.
How would you modify the input to the function if you want to find the common elements
between the columns of the matrix

```python
def common_elements_rows(A):
    common = set(A[0])
    for row in A[1:]:
        row_elements = set(row)
        common = common.intersection(row_elements)
    return sorted(common)

def common_elements_cols(A):
    cols = []
    for j in range(len(A[0])):
        col = []
        for i in range(len(A)):
            col.append(A[i][j])
        cols.append(col)
    common = common_elements_rows(cols)
    return common

A = [[7, 1, 5, 6], [2, 6, 1, 1], [6, 1, 7, 2], [6, 6, 3, 1], [5, 5, 6, 1], [3, 6, 7, 1]]

print(common_elements_rows(A))
print(common_elements_cols(A))

```

(Graded) Define a Python class, called Point, that creates an object with three data attributes
x, y, and z, and one read-only property r.
(a) The class must have an __init__ method that takes the values of the coordinates of the
point as its arguments. The value of the z coordinate is zero by default. An object of this
class can be instantiated by specifying its coordinate values as
cur_point = Point(x=1, y=2, z=3)

(b) Each object of type Point must have a read-only property called r that is equal to the
distance of the Point object from the origin. Note that a read-only property is usually
defined as a protected attribute (whose name is preceded by a single underscore) and only
has a getter method and no setter method. To show that r is a read-only property, try to
assign a new value to it.

(c) Define an appropriate method so that when an object of class Point is printed, a tuple is
displayed in the form of (x, y, z).

(d) Override the comparison ordering methods/operators so that when two objects of type
Point ae compared, their distances from the origin are compared. To that end, override
the __eq__ and __lt__ methods and then use them to define all other comparison
operators.

(e) Override the built-in addition (i.e., __add__) and addition-assignment (i.e., __iadd__)
operators so that when two objects of type Point are added, an object of type Point is
returned whose coordinates are the added values of the coordinates of the two operands.

(f) Define a method, called asdict, that returns the Point object as a dictionary. When this
method is called, a dictionary is returned in the form of {‚Äòx‚Äô:x, ‚Äòy‚Äô:y, ‚Äòz‚Äô:z}.

```python
import math 

class Point:
    def __init__(Self, x, y, z = 0):
        Self.x = x
        Self.y = y
        Self.z = z 
    
    @property
    def r(self):
        self._r = math.sqrt(self.x**2 + self.y**2 + self.z**2)
        return self._r
    
    def __str__(self):
        return f'({self.x}, {self.y}, {self.z})'
    
    def __eq__(self, other):
        return self.r == other.r
    
    def __lt__(self, other):
        return self.r < self.other
    
    def __gt__(self, other):
        return self.r > self.other

    def __iadd__(self, other):
        self.x += other.x
        self.y += other.y
        self.z += other.z
        return self
    
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y, self.z + other.z)

    def asdict(self):
        return {'x': self.x, 'y': self.y, 'z': self.z}

p = Point(1, 2, 3)
print(p)
```
